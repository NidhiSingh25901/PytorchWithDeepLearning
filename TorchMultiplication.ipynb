```python
#Multiplication of tensors using Pytorch
```

```python
import torch
```

```python
tensor1 = torch.randn(3)
tensor2 = torch.randn(3)
torch.matmul(tensor1, tensor2)
```

```python
tensor3 = torch.tensor([[1,2,3],[3,4,5]])
tensor4 = torch.tensor([[1,2],
                       [3,4],
                       [5,6]])
val = torch.matmul(tensor3, tensor4)
val
```
